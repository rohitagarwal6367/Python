```python
from pytube import YouTube

def download_video():
    # Step 1: User se YouTube video ka URL lo
    url = input("Enter the YouTube video URL: ")

    try:
        # Step 2: YouTube object banao
        yt = YouTube(url)

        # Step 3: Video ka title print karo
        print(f"\nVideo Title: {yt.title}")

        # Step 4: Available streams (qualities) show karo
        print("\nAvailable Streams (Quality):")
        streams = yt.streams.filter(progressive=True)  # Progressive streams (video + audio)
        available_streams = []
        for index, stream in enumerate(streams):
            print(f"{index + 1}. Resolution: {stream.resolution} | File Size: {round(stream.filesize / 1024 / 1024, 2)} MB")
            available_streams.append(stream)

        # Step 5: User se quality select karne ko kaho
        choice = int(input("\nSelect the stream number to download (e.g., 1, 2, etc.): "))
        if choice < 1 or choice > len(available_streams):
            print("Invalid choice! Exiting.")
            return

        selected_stream = available_streams[choice - 1]

        # Step 6: Save location ka option do (optional)
        save_path = input("\nEnter the save path (press Enter for current folder): ")
        if save_path.strip() == "":
            save_path = None

        # Step 7: Download video
        print("\nDownloading...")
        selected_stream.download(output_path=save_path)
        print("Download Complete!")

    except Exception as e:
        print(f"An error occurred: {e}")

# Call the function
if __name__ == "__main__":
    download_video()
```
