```python
from diffusers import StableDiffusionPipeline
import torch
```

```python
# Load the Stable Diffusion model from Hugging Face
model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)

# Move the model to GPU for faster processing
pipe = pipe.to("cuda")
```

```python
# Define a text prompt for image generation
prompt = "A fantasy landscape with flying islands and waterfalls"

# Generate the image based on the text prompt
image = pipe(prompt).images[0]


```

```python
# Display the image
image.show()
```
