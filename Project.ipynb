```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/student_marks.csv")
```

```python
df['Average'] = df[['Math', 'Science', 'English', 'History']].mean(axis=1)
```

```python
top_performers = df.sort_values(by='Average', ascending=False).head(3)
```

```python
subject_averages = df[['Math', 'Science', 'English', 'History']].mean()
```

```python
print("Student Averages:")
print(df[['Name', 'Average']])

print("\nTop Performers:")
print(top_performers[['Name', 'Average']])

print("\nSubject-Wise Averages:")
print(subject_averages)

print("\nSubject-Wise Total:")
print(df[['Math', 'Science', 'English', 'History']].sum())
```

```python
plt.figure(figsize=(8, 5))
subject_averages.plot(kind='bar', color=['blue', 'green', 'orange', 'red'])
plt.title('Average Marks per Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.show()
```
