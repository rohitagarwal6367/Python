```python
import yt_dlp

def download_youtube_video(url):
    try:
        # Download options
        ydl_opts = {
            'format': 'bestvideo+bestaudio/best',  # Best quality video + audio
            'outtmpl': '%(title)s.%(ext)s',       # Save with video title
        }

        # Create a downloader object
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            print("\nDownloading...")
            ydl.download([url])
            print("\nDownload Complete! Check the Files tab for the downloaded video.")

    except Exception as e:
        print(f"An error occurred: {e}")

# Input from the user
url = input("Enter the YouTube video URL: ")
download_youtube_video(url)
```

```python
def download_youtube_video_with_quality(url):
    try:
        ydl_opts = {}
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            # Retrieve video info
            info = ydl.extract_info(url, download=False)
            formats = info.get('formats', [])

            print("\nAvailable Formats:")
            for i, fmt in enumerate(formats):
                print(f"{i+1}. {fmt['format_note']} - {fmt['filesize_approx']//1024//1024 if fmt.get('filesize_approx') else 'Unknown'} MB")

            # User selects quality
            choice = int(input("\nSelect the format number to download: ")) - 1
            selected_format = formats[choice]['format_id']

            # Download the selected format
            ydl_opts = {'format': selected_format}
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                print("\nDownloading...")
                ydl.download([url])
                print("\nDownload Complete!")

    except Exception as e:
        print(f"An error occurred: {e}")

# Input from the user
url = input("Enter the YouTube video URL: ")
download_youtube_video_with_quality(url)
```

```python
import yt_dlp

def download_youtube_video_with_fallback(url):
    try:
        ydl_opts = {}
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            # Retrieve video info
            info = ydl.extract_info(url, download=False)
            formats = info.get('formats', [])

            # Display available formats
            print("\nAvailable Formats:")
            for i, fmt in enumerate(formats):
                # Flexible extraction for format details
                resolution = fmt.get('format_note', 'Unknown')
                filesize = fmt.get('filesize_approx', 'Unknown size')
                if filesize != 'Unknown size':
                    filesize = f"{filesize // 1024 // 1024} MB"
                else:
                    filesize = "Unknown size"
                format_id = fmt.get('format_id', 'Unknown')
                print(f"{i+1}. Resolution: {resolution} | Filesize: {filesize} | Format ID: {format_id}")

            # User selects format
            choice = int(input("\nSelect the format number to download (e.g., 1, 2, etc.): ")) - 1
            selected_format = formats[choice].get('format_id', None)

            if not selected_format:
                print("Invalid format selected!")
                return

            # Download selected format
            ydl_opts = {'format': selected_format}
            with yt_dlp.YoutubeDL(ydl_opts) as ydl:
                print("\nDownloading...")
                ydl.download([url])
                print("\nDownload Complete!")

    except Exception as e:
        print(f"An error occurred: {e}")

# User input
url = input("Enter the YouTube video URL: ")
download_youtube_video_with_fallback(url)
```
