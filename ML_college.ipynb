# Naive Bayes

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df=pd.read_csv("spam.csv")
```

```python
df.head()
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.Message,df.spam)
```

```python
v=CountVectorizer()
X_train_count=v.fit_transform(X_train.values)
X_train_count.toarray()[:2]
```

```python
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails=[
'Hey mohan, can we get together to watch football game tomorrow?',
'Upto 20% discount on parking, exclusive offer just for you, Dont miss this reward!'
]
emails_count=v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count=v.transform(X_test)
model.score(X_test_count,y_test)
```

# Decision Tree

```python
import pandas as pd
import math
import numpy as np
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('ID3.csv')
df.head()
```

```python
le=LabelEncoder()
for col in df.columns:
  df[col]=le.fit_transform(df[col])

print("\nEncodeed Data Set (Numerical)")
print(df)
```

```python
X=df.drop('PlayTennis',axis=1)
y=df['PlayTennis']
```

```python
model=DecisionTreeClassifier(criterion='entropy',random_state=42)
model.fit(X,y)
print("\n Decision model trained.")
```

```python
new_sample=pd.DataFrame({'Outlook':[2],'Temperature':[0],'Humidity':[0],'Wind':[1]})
prediction_encoded=model.predict(new_sample)
```

```python
target_le=LabelEncoder()
target_le.fit(df['PlayTennis'])
prediction=target_le.inverse_transform(prediction_encoded)

print(f"\nNew sample input (Encoded): \n{new_sample.iloc[0].to_dict()}")
print(f"Prediction for the new sample: {prediction[0]}")
```

```python
plt.figure(figsize=(15,11))
plot_tree(
    model,feature_names=X.columns.tolist(),
    filled=True,
    rounded=True
)
plt.title("Decision Tree using ID3 (Entropy Criterion)")
plt.show()
```

# Logistic Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# df = pd.read_csv('LGR.csv')
df=pd.DataFrame({'hours':[1,3,5,7],
                 'pass':[0,0,1,1]})
print(df)
df.to_csv("Hour_pass.csv")
```

```python
x=df[['hours']]
y=df['pass']
```

```python
model = LogisticRegression()
model.fit(x,y)
```

```python
x_new = pd.DataFrame({'hours':[5,10]})

predictions=model.predict(x_new)
probabilities=model.predict_proba(x_new)[:,1]
```

```python
print(f"Prediction for 5 hours: {'Pass' if predictions[0]==1 else'Fail'}(Prob:{probabilities[0]:.4f})")
print(f"Prediction for 10 hours: {'Pass' if predictions[1]==1 else 'Fail'}(Prob:{probabilities[1]:.4f})")
```

```python
x_range_arr=np.linspace(0,12,100).reshape(-1,1)
x_range_df=pd.DataFrame(x_range_arr,columns=['hours'])
y_range_prob=model.predict_proba(x_range_df)[:,1]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x,y,color='red',label="Actual Date", zorder=5)
plt.plot(x_range_df['hours'],y_range_prob,color='blue',label='Logistic Regression Curve')
plt.axhline(0.5,color='black',linestyle='--',alpha=0.5,label='Decision Threshold (0.5)')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Logistic Regression (Resolved Warings)')
plt.legend()
plt.grid(True,alpha=0.3)
plt.show()
```
