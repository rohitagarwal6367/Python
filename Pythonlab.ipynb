```python
# Linear Search
```

```python
n = int(input())
arr = list(map(int, input().split()))
target = int(input())

index = -1
for i in range(n):
    if arr[i] == target:
        index = i
        break

print(index)
```

```python
# Binary Search
n = int(input())
arr = list(map(int, input().split()))
target = int(input())

low = 0
high = n - 1
ans = -1

while low <= high:
    mid = (low + high) // 2

    if arr[mid] == target:
        ans = mid
        break
    elif arr[mid] < target:
        low = mid + 1
    else:
        high = mid - 1

print(ans)
```

```python
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):

        for j in range(0, n - i - 1):

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]


numbers = [64, 34, 25, 12, 22, 11, 90]

bubble_sort(numbers)

print("Sorted array:", numbers)
```
