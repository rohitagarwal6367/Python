```python
import numpy as np
import cv2
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
```

```python
# Simulating with MNIST as placeholder for food dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Normalize the data
x_train = x_train / 255.0
x_test = x_test / 255.0

# Convert labels to categorical
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)
```

```python
def apply_filters(img):
    # Resize and apply GaussianBlur + Sobel Edge Detection
    img = cv2.resize(img, (28, 28))  # standard size
    blurred = cv2.GaussianBlur(img, (3, 3), 0)
    sobelx = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)
    edge_img = cv2.magnitude(sobelx, sobely)
    return edge_img

# Apply filter on training data
x_train_filtered = np.array([apply_filters(img) for img in x_train])
x_test_filtered = np.array([apply_filters(img) for img in x_test])
```

```python
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')  # change 10 to number of food classes
])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
model.fit(x_train_filtered, y_train_cat, epochs=5, validation_data=(x_test_filtered, y_test_cat))
```

```python
test_image = x_test[0]
filtered_image = apply_filters(test_image)
prediction = model.predict(np.expand_dims(filtered_image, axis=0))
print("Predicted class:", np.argmax(prediction))

plt.imshow(test_image, cmap='gray')
plt.title("Original Test Image")
plt.show()
```
