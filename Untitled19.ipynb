```python
import os
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt

```

```python
# इमेज साइज
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# डेटा ऑगमेंटेशन
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.7, 1.3],
    fill_mode='nearest',
    validation_split=0.2  # 20% डेटा वैलिडेशन के लिए
)

# ट्रेन और वैलिडेशन जनरेटर
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/archive1/Indian Food Images/Indian Food Images',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training')

validation_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/archive1/Indian Food Images/Indian Food Images',
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation')

# क्लासेस प्रिंट करें
print("क्लासेस: ", train_generator.class_indices)
```

```python
def build_model(num_classes):
    # बेस मॉडल (EfficientNetB0)
    base_model = EfficientNetB0(
        weights='imagenet',
        include_top=False,
        input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

    # फाइन-ट्यूनिंग के लिए कुछ लेयर्स खोलें
    base_model.trainable = True
    for layer in base_model.layers[:-20]:
        layer.trainable = False

    # नया मॉडल बनाएं
    model = models.Sequential([
        base_model,
        layers.GlobalAveragePooling2D(),
        layers.BatchNormalization(),
        layers.Dense(1024, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(512, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax')
    ])

    # मॉडल को कंपाइल करें
    model.compile(
        optimizer=Adam(learning_rate=1e-4),
        loss='categorical_crossentropy',
        metrics=['accuracy'])

    return model

# मॉडल बनाएं
model = build_model(train_generator.num_classes)
model.summary()
```

```python
# क्लास वेट्स (अगर डेटा असंतुलित हो)
from sklearn.utils import class_weight
class_weights = class_weight.compute_class_weight(
    'balanced',
    classes=np.unique(train_generator.classes),
    y=train_generator.classes)
class_weights = dict(enumerate(class_weights))

# कॉलबैक्स
callbacks = [
    EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)
]

# एपोक्स की संख्या
EPOCHS = 50
```

```python
# ट्रेनिंग शुरू करें
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    validation_data=validation_generator,
    validation_steps=len(validation_generator),
    epochs=EPOCHS,
    callbacks=callbacks,
    class_weight=class_weights
)
```

```python
# सेल 1: EarlyStopping को एडजस्ट करें
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(
    monitor='val_accuracy',  # val_loss की जगह val_accuracy पर नजर रखें
    patience=20,            # 20 epochs तक इंतजार करें
    min_delta=0.001,        # मिनिमम सुधार थ्रेशोल्ड
    mode='max',             # accuracy को maximize करना है
    restore_best_weights=True
)
```

```python
# सेल 2: Optimizer और Learning Rate
from tensorflow.keras.optimizers import Adam

optimizer = Adam(
    learning_rate=0.0001,   # LR को 0.0001 पर सेट करें
    clipnorm=1.0            # ग्रेडिएंट क्लिपिंग जोड़ें
)
```

```python
# सेल 3: मॉडल कंपाइलेशन
model.compile(
    optimizer=optimizer,
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# सेल 4: डेटा जनरेटर
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    brightness_range=[0.8,1.2],  # कम एक्सट्रीम वैल्यू
    fill_mode='nearest',
    validation_split=0.2
)

# जनरेटर्स को रीइनिशियलाइज़ करें (अपने डेटा पाथ के साथ)
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/archive1/Indian Food Images/Indian Food Images',
    target_size=(224, 224),
    batch_size=64,  # बैच साइज बढ़ाया
    class_mode='categorical',
    subset='training'
)
```

```python
# सेल 5: ट्रेनिंग रन
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    validation_data=validation_generator,
    validation_steps=len(validation_generator),
    epochs=100,  # ज्यादा epochs
    callbacks=[early_stop],
    verbose=1
)
```
